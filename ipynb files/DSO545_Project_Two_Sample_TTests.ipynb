```python
# FG%

# Dropping NAN values
X = Without_TOT_D1.dropna(subset=['FG%'])
Y = Without_TOT_D2.dropna(subset=['FG%'])

# Checking correlation
correlation = X['FG%'].corr(Y['FG%'])
correlation

print(f"Correlation between Decade 1 FG% and Decade 2 FG%: {correlation:.4f}")

# Checking means
print(f"Average FG% of Decade 1: {X['FG%'].mean():.2f}")
print(f"Average FG% of Decade 2: {Y['FG%'].mean():.2f}")

# T-Test
t_stat, p_value = ttest_ind(Y['FG%'], X['FG%'], alternative = "greater")

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.2f}")

# Confidence Interval
cm = sms.CompareMeans(sms.DescrStatsW(Y['FG%']), sms.DescrStatsW(X['FG%']))
conf_interval = cm.tconfint_diff(alpha=0.05, usevar='unequal')  # 95% Confidence Interval

print(f"95% Confidence Interval for the difference in means: ({conf_interval[0]:.4f}, {conf_interval[1]:.4f})")
```
