# HEATMAP AND VIF FOR WITHOUT_TOT

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
#VIF CALCULATION CODE
X = Without_TOT.drop(columns=['PTS', 'Player', 'Pos', 'Tm', 'Season', 'FG', 'FG%', 'FTA', 'FT', '2PA', '2P', '3P', '3P%', 'ORB', 'DRB'])  # Replace 'dependent_variable_column' with the actual dependent variable column
X = X.replace([np.inf, -np.inf], np.nan).dropna()


# Add a constant term for the intercept
X = sm.add_constant(X)

# Calculate VIF for each predictor
vif = pd.DataFrame()
vif["Variable"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif)

#HEATMAP CORR MATRIX CODE
correlation_matrix = Without_TOT.corr()
# Visualizing the correlation matrix using a heatmap
plt.figure(figsize=(20,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

# REGRESSIONS

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
from sklearn.impute import SimpleImputer

# Identify numeric and non-numeric columns
numeric_cols = Without_TOT.select_dtypes(include=[np.number]).columns
non_numeric_cols = Without_TOT.select_dtypes(exclude=[np.number]).columns

# Impute missing values with mean for numeric columns and most frequent for non-numeric columns
numeric_imputer = SimpleImputer(strategy='mean')
non_numeric_imputer = SimpleImputer(strategy='most_frequent')

Without_TOT[numeric_cols] = numeric_imputer.fit_transform(Without_TOT[numeric_cols])
Without_TOT[non_numeric_cols] = non_numeric_imputer.fit_transform(Without_TOT[non_numeric_cols])

# Define features (X) and target variable (y)
x = Without_TOT[['G', 'FGA', '3PA', '2P%', 'FT%', 'TRB', 'AST', 'STL', 'BLK', 'PF']]
y = Without_TOT['PTS']

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Create and fit the linear regression model to the training set
model = LinearRegression()
model.fit(x_train, y_train)

# Make predictions on both the training and testing sets
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)

# Calculate and print metrics for both training and testing sets
mse_train = mean_squared_error(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)
rmse_train = np.sqrt(mse_train)
rmse_test = np.sqrt(mse_test)
mae_train = mean_absolute_error(y_train, y_train_pred)
mae_test = mean_absolute_error(y_test, y_test_pred)
r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_test_pred)

print("Training Metrics:")
print(f'Mean Squared Error (MSE): {mse_train:.2f}')
print(f'Root Mean Squared Error (RMSE): {rmse_train:.2f}')
print(f'Mean Absolute Error (MAE): {mae_train:.2f}')
print(f'R-squared: {r2_train:.2f}\n')

print("Testing Metrics:")
print(f'Mean Squared Error (MSE): {mse_test:.2f}')
print(f'Root Mean Squared Error (RMSE): {rmse_test:.2f}')
print(f'Mean Absolute Error (MAE): {mae_test:.2f}')
print(f'R-squared: {r2_test:.2f}')
```
